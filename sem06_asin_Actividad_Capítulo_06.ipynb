```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.

"""
    Función para buscar un elemento en una lista ordenada utilizando el algoritmo de búsqueda binaria recursiva.

    Parámetros:
        - nums: Lista ordenada de números.
        - x: Elemento a buscar.
        - low: Índice inferior de la lista.
        - high: Índice superior de la lista.

    Retorna:
        - La posición del elemento si se encuentra en la lista, de lo contrario retorna -1.
    """

"""La función se llama a sí misma en cada iteración para buscar en la mitad izquierda o derecha de la lista
 dependiendo de la comparación entre el elemento medio y el objetivo. Este enfoque divide eficientemente 
 la lista en cada iteración, reduciendo a la mitad el espacio de búsqueda en cada paso."""

 
def bin_search2(nums, x, low, high):

    if low > high:  # Si el índice inferior es mayor que el superior, la lista está vacía
        return -1
    else:
        mid = (low + high) // 2  # Calcula el índice medio de la lista
        if nums[mid] == x:  # Si el elemento en la posición media es igual al objetivo, se ha encontrado
            return mid
        elif nums[mid] > x:  # Si el elemento medio es mayor que el objetivo, buscar en la mitad izquierda
            return bin_search2(nums, x, low, mid - 1)
        else:  # Si el elemento medio es menor que el objetivo, buscar en la mitad derecha
            return bin_search2(nums, x, mid + 1, high)


nums = [15, 10, 12, 20, 25, 13, 22]  # Lista de números ordenados
target = 20  # Elemento a buscar
low = 0
high = len(nums) - 1

result = bin_search2(nums, target, low, high)  # Llamada a la función de búsqueda binaria
if result != -1:
    print(f"El elemento {target} se encuentra en la posición {result}.")
else:
    print(f"El elemento {target} no se encuentra en la lista.")


```

### La diferencia entre los dos enfoques:

1- Una de las diferencias entre el enfoque divide y venceras es que no exite la superposicion de problemas y cada problema es independiente y unico,
cosa que en la programación dinamica es todo lo contrario, allí si se necesita solapar subproblemas para llegar a una solucion optima.

2- La segunda diferencia encontrada es la memorizacion de resultados, en la programacion dinamica es más común esta tecnica, asi no se recalculan subproblemas antes calculados.
En la metodologia divide y venceras no existe la memorizacion debido a que cada uno de los subproblemas son unicos y independientes.


### Casos donde fuciona mejor la programacion dinamica:


1.Problemas de optimización donde el problema puede dividirse en subproblemas que se solapan.

2.Problemas de secuenciación donde la solución óptima involucra tomar decisiones en un orden específico.

3.Problemas de caminos más cortos o más largos en grafos, donde el camino óptimo implica tomar decisiones que pueden superponerse, como encontrar el camino más corto en un grafo ponderado.
