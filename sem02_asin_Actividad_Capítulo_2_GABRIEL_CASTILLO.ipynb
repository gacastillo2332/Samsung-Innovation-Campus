```python

# Actividad: Revisión y manejo de listas bidimensionales.

## Instrucciones

1. Describa brevemente que entiende por lista bidimensional.
2. Crear una lista bidimensional.
3. Generar un doble loop para acceder de la lista bidimensional
4. Sumar todo los elementos de esta lista bidimensional.

````python
# Espacio para desarrollar el código, escriba comentarias breves para explicar su funcionamiento.

'''Lo que entiendo por arreglo bidimencional es una estructura de dato que tienen filas 
y columnas y que es generalmente representado por una malla rectangular.'''


numeros = [[1,2,3],[4,5,6],[7,8,9]]
suma = 0

for i in  range(0,3,1):
    for j in  range(0,3,1):
        suma = suma + numeros[i][j]

print("El resultado de la suma de todos los elementos de la matriz es: {}".format(suma))
````

# Actividad: Revision de los métodos usados en el manejo de diccionarios.

## Instrucciones

1. Crear un diccionario.
2. Testear los siguientes métodos para el majeo de diccionarios get(), keys(), values(), items(), pop(), clear()
3. Agregar un breve comentario sobre su funcionamiento.

````python
# Espacio para desarrollar el código.


persona = {"dni":"8-983-725","nombre":"Gabriel","apellido":"Castillo","edad":21}

#La funcion get() retorna el valor asociado con la clave especificada.
print("El valor asociado a la key nombre es: ", persona.get("nombre"))


#La funcion keys() retorna el nombre de todas las claves existentes en nuestro diccionario.
print("Valor retornado por la funcion keys(): ", persona.keys())

#La funcion values() retorna todos los valores que estan guardados en el diccionario.
print("Valor retornado por la funcion values(): ", persona.values())

#La funcion items() retorna una vista de tuplas (clave, valor) para cada par clave-valor en nuestro diccionario.
print("Valor retornado por la funcion items(): ", persona.items())

#La funcion pop() como parametro toma una keys y elimina la clave especificada y devuelve su valor asociado.

print("Valor retornado por la funcion pop() borrando la keys edad: ", persona.pop("edad"))

#La funcion clear() elimina todos los elementos del diccionario dejándolo vacío.
persona.clear()
print("Este es el diccionario luego de aplicar la funcion clear(): ",persona.items())
````
